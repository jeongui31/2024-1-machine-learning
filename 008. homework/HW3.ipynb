```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN

df = pd.read_csv("urbanGB-simple.csv")
```

```python
df.head()
```

```python
sc = StandardScaler()
X = sc.fit_transform(df)
```

```python
distortions = []

for i in range(1, 11):
    km = KMeans(n_clusters=i, 
                n_init = 10,
                init='k-means++',
                max_iter=300,
                random_state=0)
    km.fit(X)
    distortions.append(km.inertia_)
    
plt.plot(range(1, 11), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.tight_layout()

plt.show()
```

```python
silhouette_scores = []

for i in range(2, 11):
    km = KMeans(n_clusters=i,
               n_init=10,
               init='k-means++',
               max_iter=300,
               random_state=0)
    
    y_km = km.fit_predict(X)
    silhouette_scores.append(silhouette_score(X, y_km, metric='euclidean'))
    
plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette score')
plt.tight_layout()

plt.show()
```

```python
km = KMeans(n_clusters=4,
            n_init=10,
            init='k-means++',
            max_iter=300,
            random_state=0)

y_km = km.fit_predict(X)
y_km
```

```python
df['kmeans_cluster'] = y_km

plt.figure(figsize=(10, 8))
plt.scatter(df.iloc[:, 0], df.iloc[:, 1], c=df['kmeans_cluster'], cmap='viridis', marker='o', edgecolor='k', s=50)
plt.title('K-means Clustering (k=4)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
db = DBSCAN(eps=0.32, min_samples=16,metric='euclidean')

y_db = db.fit_predict(X)

df['dbscan_cluster'] = y_db
outliers = df[df['dbscan_cluster'] == -1]

# DBSCAN   
plt.figure(figsize=(10, 8))
plt.scatter(df.iloc[:, 0], df.iloc[:, 1], c=df['dbscan_cluster'], cmap='viridis', marker='o', edgecolor='k', s=50)
plt.scatter(outliers.iloc[:, 0], outliers.iloc[:, 1], c='red', marker='x', edgecolor='k', s=50, label='Outliers')
plt.title('DBSCAN clustering')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Dictionary to store the number of clusters for each parameter combination
cluster_counts = {}

# Find the number of clusters for various eps and min_samples values
for eps in np.arange(0.1, 1.0, 0.05):
    for min_samples in range(5, 50, 5):
        db = DBSCAN(eps=eps, min_samples=min_samples, metric='euclidean')
        y_db = db.fit_predict(X)
        n_clusters = len(set(y_db)) - (1 if -1 in y_db else 0)
        cluster_counts[(eps, min_samples)] = n_clusters

# Convert the dictionary to a DataFrame for easier plotting
cluster_counts_df = pd.DataFrame.from_dict(cluster_counts, orient='index', columns=['n_clusters'])
cluster_counts_df.index = pd.MultiIndex.from_tuples(cluster_counts_df.index, names=['eps', 'min_samples'])
cluster_counts_df = cluster_counts_df.reset_index()

# Plot the number of clusters as a function of eps and min_samples
fig, ax = plt.subplots(figsize=(12, 8))
for min_samples in cluster_counts_df['min_samples'].unique():
    subset = cluster_counts_df[cluster_counts_df['min_samples'] == min_samples]
    ax.plot(subset['eps'], subset['n_clusters'], marker='o', label=f'min_samples={min_samples}')

ax.set_title('Number of Clusters as a Function of eps and min_samples')
ax.set_xlabel('eps')
ax.set_ylabel('Number of Clusters')
ax.legend(title='min_samples')
plt.grid(True)
plt.show()
```

```python
import numpy as np
from sklearn.datasets import fetch_20newsgroups
from sklearn.model_selection import train_test_split
import re
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
categories = ['alt.atheism', 'soc.religion.christian',
             'comp.graphics', 'sci.med']

news = fetch_20newsgroups(categories=categories,
                         shuffle=False,
                         random_state=42)

X=news.data
y=news.target

X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                   test_size=0.3,
                                                   stratify=y, 
                                                   random_state=1)
```

```python
def preprocessor(text):
    text = re.sub('[\W]+', ' ', text)
    text = text.lower()
    return text

stemmer = PorterStemmer()

def tokenizer_stemmer(text):
    text_tokens = word_tokenize(text)
    return [stemmer.stem(word) for word in text_tokens]
```

```python
stop = stopwords.words('english')

tfidf = TfidfVectorizer(
    preprocessor=preprocessor,
    tokenizer=tokenizer_stemmer,
    stop_words=stop,
    max_df=0.1,
    min_df=10
)

X_train_tfidf = tfidf.fit_transform(X_train)
X_test_tfidf = tfidf.transform(X_test)
```

```python
lr = LogisticRegression(penalty='l2')
lr.fit(X_train_tfidf, y_train)

print(lr.score(X_train_tfidf, y_train))
print(lr.score(X_test_tfidf, y_test))
```

```python
tree = DecisionTreeClassifier(max_depth=20)
tree.fit(X_train_tfidf, y_train)

print(tree.score(X_train_tfidf, y_train))
print(tree.score(X_test_tfidf, y_test))
```

```python
# finding most important terms for sentiment analysis of movie review
importances = tree.feature_importances_
indices = np.argsort(importances)[::-1]

for f in range(10):
    print("%2d. %-30s %f" % (f+1, 
                             [w for w, n in tfidf.vocabulary_.items() if n == indices[f]],
                             importances[indices[f]]))
```

```python
new_sentences = [
    'The outbreak was declared a global pandemic by the World Health Organization (WHO) on 11 March.',
    'Today, computer graphics is a core technology in digitalphotography, film, video games, cell phone and computer displays,and many specialized applications.',
    'Arguments for atheism range from philosophical to social and historical approaches.',
    'The Bible is a compilation of many shorter books written at different times by a variety of authors, and later assembled into the biblical canon.'
]

new_sentences_tfidf = tfidf.transform(new_sentences)

lr_predictions = lr.predict(new_sentences_tfidf)
tree_predictions = tree.predict(new_sentences_tfidf)

print("\nLogistic Regression Predictions:")
for sentence, category in zip(new_sentences, lr_predictions):
    print(f"'{sentence}' => {news.target_names[category]}")

print("\nDecision Tree Predictions:")
for sentence, category in zip(new_sentences, tree_predictions):
    print(f"'{sentence}' => {news.target_names[category]}")
```
